### 1. **Jika model linear regression atau decision tree mengalami underfitting pada dataset ini, strategi apa yang akan digunakan untuk meningkatkan performanya? Bandingkan setidaknya dua pendekatan berbeda (misal: transformasi fitur, penambahan features, atau perubahan model ke algoritma yang lebih kompleks), dan jelaskan bagaimana setiap solusi memengaruhi bias-variance tradeoff!**

#### Pendekatan 1: **Penambahan Fitur (Feature Engineering)**

* **Deskripsi**: Menambahkan fitur baru yang relevan dengan target atau mengubah representasi fitur yang ada (misalnya menggunakan Polynomial Features untuk menangkap hubungan non-linear). Hal ini bisa memperbaiki performa model dengan memberikan lebih banyak informasi untuk memprediksi target.
* **Bias-Variance Tradeoff**: Penambahan fitur baru biasanya akan mengurangi bias model (lebih fleksibel dalam menangkap pola) namun bisa meningkatkan varians (menjadi lebih mudah overfitting). Oleh karena itu, kita perlu memastikan fitur yang ditambahkan benar-benar relevan.

#### Pendekatan 2: **Menggunakan Model yang Lebih Kompleks (misalnya, Random Forest atau Gradient Boosting)**

* **Deskripsi**: Jika linear regression atau decision tree terlalu sederhana dan mengalami underfitting, kita dapat beralih ke model yang lebih kompleks, seperti Random Forest atau Gradient Boosting, yang lebih kuat dalam menangani hubungan non-linear dan interaksi antar fitur.
* **Bias-Variance Tradeoff**: Menggunakan model yang lebih kompleks dapat mengurangi bias secara signifikan (lebih fleksibel untuk menangkap pola data yang lebih rumit). Namun, model yang lebih kompleks juga cenderung meningkatkan varians, dan mungkin lebih rentan terhadap overfitting jika tidak dikendalikan dengan baik melalui teknik regulasi (seperti pruning pada decision tree atau menggunakan hyperparameter tertentu pada ensemble methods).

### 2. **Selain MSE, jelaskan dua alternatif loss function untuk masalah regresi (misal: MAE, Huber loss) dan bandingkan keunggulan serta kelemahannya. Dalam skenario apa setiap loss function lebih cocok digunakan?**

#### Alternatif 1: **MAE (Mean Absolute Error)**

* **Keunggulan**: MAE menghitung selisih absolut antara nilai yang diprediksi dan nilai aktual, yang lebih robust terhadap outlier dibandingkan MSE.
* **Kelemahan**: MAE kurang sensitif terhadap perbedaan besar antara prediksi dan nilai aktual (sehingga bisa lebih sulit untuk memberikan penalti yang sesuai pada model).
* **Kapan digunakan**: MAE lebih cocok digunakan ketika data mengandung outlier, atau ketika kita ingin memperlakukan kesalahan besar dan kecil secara setara (karena tidak mengkuadratkan kesalahan).

#### Alternatif 2: **Huber Loss**

* **Keunggulan**: Huber loss menggabungkan sifat baik dari MSE dan MAE. Untuk kesalahan kecil, ia bertindak seperti MSE (menghukum kesalahan kecil dengan kuadrat), tetapi untuk kesalahan besar, ia bertindak lebih seperti MAE (menghindari penalti yang terlalu besar pada outlier).
* **Kelemahan**: Huber loss memiliki parameter (delta) yang perlu diatur, dan pemilihan delta yang tidak tepat dapat mengarah pada hasil yang kurang optimal.
* **Kapan digunakan**: Huber loss sangat berguna ketika dataset mengandung outlier, namun kita tidak ingin kesalahan kecil dihukum terlalu berat seperti pada MSE. Ini memberikan solusi yang lebih baik ketika data tidak terdistribusi dengan baik atau memiliki noise.

### 3. **Tanpa mengetahui nama fitur, metode apa yang dapat digunakan untuk mengukur pentingnya setiap fitur dalam model? Jelaskan prinsip teknikal di balik metode tersebut (misal: koefisien regresi, feature importance berdasarkan impurity reduction) serta keterbatasannya!**

#### Metode 1: **Koefisien Regresi (untuk Linear Regression)**

* **Prinsip Teknis**: Dalam regresi linear, koefisien dari setiap fitur menunjukkan pengaruh langsungnya terhadap target. Fitur dengan koefisien yang lebih besar memiliki dampak yang lebih besar pada prediksi model.
* **Keterbatasan**: Koefisien regresi hanya berlaku untuk model linear dan dapat terdistorsi oleh multikolinearitas (hubungan antara fitur yang saling berkorelasi). Fitur yang terkorleasi mungkin tampak tidak penting meskipun sebenarnya mereka membawa informasi penting.

#### Metode 2: **Feature Importance Berdasarkan Impurity Reduction (untuk Decision Tree dan Ensemble Methods)**

* **Prinsip Teknis**: Decision trees (dan model ensemble seperti Random Forest atau Gradient Boosting) mengukur pentingnya fitur dengan melihat bagaimana fitur tersebut mengurangi impurity dalam pemisahan data. Fitur yang sering digunakan untuk membagi data pada simpul yang mengurangi impurity lebih banyak dianggap lebih penting.
* **Keterbatasan**: Feature importance pada decision tree tidak selalu memberikan penilaian yang adil untuk fitur yang memiliki banyak kategori atau kategori yang sangat tidak seimbang. Selain itu, model ini dapat terpengaruh oleh overfitting, terutama pada pohon keputusan yang sangat dalam.

### 4. **Bagaimana mendesain eksperimen untuk memilih hyperparameter optimal (misal: learning rate untuk SGDRegressor, max\_depth untuk Decision Tree) pada dataset ini? Sertakan analisis tradeoff antara komputasi, stabilitas pelatihan, dan generalisasi model!**

#### Desain Eksperimen:

1. **Pemilihan Hyperparameter**:

   * Untuk model seperti SGDRegressor, kita bisa menyesuaikan hyperparameter seperti learning rate, n\_iter, dan penalty.
   * Untuk Decision Tree, kita bisa menyesuaikan max\_depth, min\_samples\_split, dan min\_samples\_leaf.

2. **Grid Search vs Random Search**:

   * **Grid Search**: Mencoba semua kombinasi hyperparameter dalam rentang yang ditentukan. Lebih lengkap, tetapi memerlukan banyak waktu komputasi.
   * **Random Search**: Memilih kombinasi acak dari hyperparameter, lebih efisien dalam hal komputasi namun dapat melewatkan kombinasi optimal.

3. **K-fold Cross Validation**:

   * Untuk mengukur kinerja model, kita dapat menggunakan K-fold cross-validation untuk memastikan hasil yang lebih stabil dan menghindari overfitting pada data training.

#### Trade-off:

* **Komputasi**: Grid search lebih mahal dalam hal waktu komputasi karena mencoba semua kombinasi. Random search lebih efisien, tetapi bisa jadi tidak mendapatkan kombinasi terbaik.
* **Stabilitas Pelatihan**: Menggunakan cross-validation memastikan bahwa model tidak terpengaruh oleh fluktuasi hasil pada set data yang berbeda.
* **Generalisasi Model**: Pengaturan hyperparameter yang tepat membantu model untuk generalisasi dengan lebih baik, menghindari overfitting dan underfitting.

### 5. **Jika menggunakan model linear regression dan residual plot menunjukkan pola non-linear serta heteroskedastisitas, langkah-langkah apa yang akan diambil? (contohnya: Transformasi data/ubah model yang akan dipakai/etc)**

#### Langkah-langkah:

1. **Transformasi Data**:

   * Jika terdapat pola non-linear, kita dapat mencoba **transformasi logaritma** atau **polynomial features** untuk menangkap hubungan yang lebih kompleks.
   * Transformasi target atau fitur bisa membantu mengurangi heteroskedastisitas dan membuat hubungan lebih linier.

2. **Ubah Model yang Digunakan**:

   * Jika model regresi linear tetap menunjukkan masalah, kita dapat beralih ke **Decision Tree**, **Random Forest**, atau **Gradient Boosting**, yang lebih fleksibel dalam menangkap hubungan non-linear.

3. **Penerapan Heteroskedastisitas**:

   * Untuk mengatasi heteroskedastisitas, kita bisa menggunakan **Weighted Least Squares (WLS)** yang memberikan bobot lebih besar pada observasi dengan variansi yang lebih kecil.
