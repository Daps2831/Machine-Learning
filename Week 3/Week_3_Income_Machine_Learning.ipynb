### Penjelasan Setiap Persamaan Matematika

Untuk memahami persamaan ini, kita perlu mendefinisikan empat komponen dasar dari **Confusion Matrix**:

* **True Positive (TP):** Prediksi Benar Positif.
    * Kasus: Model memprediksi `>50K` dan kenyataannya memang `>50K`.
* **True Negative (TN):** Prediksi Benar Negatif.
    * Kasus: Model memprediksi `<=50K` dan kenyataannya memang `<=50K`.
* **False Positive (FP):** Prediksi Salah Positif (Error Tipe I).
    * Kasus: Model memprediksi `>50K`, tetapi kenyataannya adalah `<=50K`.
* **False Negative (FN):** Prediksi Salah Negatif (Error Tipe II).
    * Kasus: Model memprediksi `<=50K`, tetapi kenyataannya adalah `>50K`.

Berikut adalah penjelasan untuk setiap metrik:

#### 1. Akurasi (Accuracy)
Mengukur proporsi total prediksi yang benar (baik positif maupun negatif) dari keseluruhan data.

**Persamaan:**
$$\text{Akurasi} = \frac{TP + TN}{TP + TN + FP + FN}$$

**Penjelasan:** (Jumlah semua tebakan yang benar) dibagi dengan (Total seluruh data). Metrik ini baik untuk dataset seimbang, tetapi menyesatkan untuk dataset tidak seimbang.

#### 2. Presisi (Precision)
Mengukur akurasi dari prediksi positif. Dari semua yang diprediksi sebagai `>50K`, berapa persen yang benar-benar `>50K`?

**Persamaan:**
$$\text{Presisi} = \frac{TP}{TP + FP}$$

**Penjelasan:** Berguna ketika biaya dari *False Positive* tinggi. Misalnya, Anda tidak ingin salah menawarkan produk premium (FP) kepada pelanggan yang sebenarnya tidak memenuhi kriteria.

#### 3. Recall (Sensitivity atau True Positive Rate)
Mengukur kemampuan model untuk "menemukan" semua kasus positif yang ada. Dari semua yang kenyataannya `>50K`, berapa persen yang berhasil diprediksi dengan benar oleh model?

**Persamaan:**
$$\text{Recall} = \frac{TP}{TP + FN}$$

**Penjelasan:** Berguna ketika biaya dari *False Negative* tinggi. Misalnya, dalam diagnosis medis, Anda tidak ingin melewatkan pasien yang sebenarnya sakit (FN).

#### 4. F1-Score
Rata-rata harmonik (harmonic mean) dari Presisi dan Recall. Metrik ini mencari keseimbangan antara keduanya.

**Persamaan:**
$$\text{F1-Score} = 2 \times \frac{\text{Presisi} \times \text{Recall}}{\text{Presisi} + \text{Recall}}$$

**Penjelasan:** F1-Score akan tinggi hanya jika Presisi dan Recall sama-sama tinggi. Ini adalah metrik yang jauh lebih baik daripada akurasi pada dataset tidak seimbang.

#### 5. AUC (Area Under the Curve)
AUC tidak memiliki formula sederhana seperti di atas. Ini adalah area di bawah **Kurva ROC (Receiver Operating Characteristic)**.

**Penjelasan Konseptual:**
* Kurva ROC memplot **Recall (True Positive Rate)** melawan **False Positive Rate** pada berbagai ambang batas klasifikasi.
* **AUC** mengukur keseluruhan area di bawah kurva ini.
    * **AUC = 1.0:** Model sempurna dalam membedakan kelas.
    * **AUC = 0.5:** Model tidak lebih baik dari tebakan acak (garis diagonal pada plot ROC).
    * **AUC < 0.5:** Model lebih buruk dari tebakan acak.
* Secara intuitif, AUC merepresentasikan probabilitas bahwa model akan memberikan skor yang lebih tinggi untuk sampel positif yang dipilih secara acak daripada sampel negatif yang dipilih secara acak.
