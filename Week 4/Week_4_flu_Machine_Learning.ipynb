Hapus Baris Dimana Target (Fatigue) Tidak Diketahui
Ini adalah langkah pertama yang tidak bisa ditawar. Kita tidak bisa melatih model jika kita tidak tahu jawabannya (apakah pasien mengalami Fatigue atau tidak).

```python
# Tampilkan jumlah baris sebelum pembersihan
print(f"Jumlah baris sebelum menangani missing values di target: {len(df)}")

# Hapus semua baris di mana kolom 'Fatigue' kosong (NaN)
df_cleaned = df.dropna(subset=['Fatigue'])

# Tampilkan jumlah baris setelah pembersihan
print(f"Jumlah baris setelah menghapus 'Fatigue' yang kosong: {len(df_cleaned)}")
```

Hapus Kolom dengan Terlalu Banyak Nilai Hilang
Setelah kita memiliki data yang targetnya valid, mari kita lihat kembali kolom-kolom fitur. Banyak kolom seperti InitialPCRDiagnosis, neutrophil, ShortnessOfBreath yang kehilangan lebih dari 90% datanya. Mengisi nilai-nilai ini akan lebih banyak menghasilkan "tebakan" daripada data asli dan bisa merusak performa model.

Strategi terbaik adalah menghapus kolom yang persentase data hilangnya sangat tinggi. Mari kita tetapkan ambang batas: kita akan menghapus kolom jika lebih dari 60% datanya hilang.

```python
# Hitung ulang nilai hilang pada data yang sudah bersih
missing_values_after_cleaning = df_cleaned.isnull().sum()
print("--- Nilai Hilang pada Data Setelah Target Dibersihkan ---")
print(missing_values_after_cleaning)

# Tentukan ambang batas untuk menghapus kolom (40% data harus ada)
threshold = len(df_cleaned) * 0.4

# Hapus kolom yang tidak memenuhi ambang batas
df_cleaned = df_cleaned.dropna(axis=1, thresh=threshold)

print("\n--- Kolom yang Tersisa Setelah Menghapus Kolom Kosong ---")
print(df_cleaned.columns.tolist())
```

 Isi Sisa Nilai yang Hilang (Imputasi)
Sekarang kita memiliki kolom-kolom yang lebih "sehat". Namun, mungkin masih ada beberapa nilai yang hilang di kolom seperti Age, Sex, Fever, dll. Kita akan mengisinya dengan strategi berikut:

Kolom Numerikal: Isi dengan nilai median (nilai tengah, lebih aman dari outlier dibanding rata-rata).
Kolom Kategorikal: Isi dengan nilai modus (nilai yang paling sering muncul).



```python
from sklearn.impute import SimpleImputer

# Pisahkan lagi kolom numerikal dan kategorikal dari dataframe yang sudah lebih bersih
# Catatan: 'Age' dan 'Temperature' mungkin numerikal. Sisanya kategorikal.
# Periksa df_cleaned.info() jika Anda tidak yakin
numerical_cols = df_cleaned.select_dtypes(include=['number']).columns
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns

# Hapus kolom target dari daftar kolom kategorikal agar tidak ter-imputasi
categorical_cols = categorical_cols.drop('Fatigue')

# Imputasi untuk data numerikal
imputer_numerical = SimpleImputer(strategy='median')
df_cleaned[numerical_cols] = imputer_numerical.fit_transform(df_cleaned[numerical_cols])

# Imputasi untuk data kategorikal
imputer_categorical = SimpleImputer(strategy='most_frequent')
df_cleaned[categorical_cols] = imputer_categorical.fit_transform(df_cleaned[categorical_cols])

# Pengecekan terakhir untuk memastikan tidak ada lagi nilai yang hilang
print("\n--- Pengecekan Akhir Nilai Hilang ---")
print(df_cleaned.isnull().sum())

print(f"Jumlah baris sekarang: {len(df_cleaned)}")
```

```python
# prompt: buatkan kode analisis distribusi kolom kategorikal

import matplotlib.pyplot as plt
import seaborn as sns

# Analisis distribusi kolom kategorikal
print("\n--- Analisis Distribusi Kolom Kategorikal ---")
for col in categorical_cols:
    print(f"\nDistribusi Kolom: {col}")
    print(df_cleaned[col].value_counts())
    print(f"\nProporsi Distribusi Kolom: {col}")
    print(df_cleaned[col].value_counts(normalize=True) * 100)
```

Pra-Pemrosesan Akhir (Encoding)
Model memerlukan semua input dalam bentuk angka. Kita akan mengubah kolom-kolom kategorikal (seperti Diagnosis, Sex, Fever, dll.) menjadi representasi numerik.

```python
# Pastikan Anda bekerja dengan dataframe yang sudah bersih, yaitu df_cleaned

# 1. Pisahkan antara Fitur (X) dan Target (y)
X = df_cleaned.drop('Fatigue', axis=1)
y = df_cleaned['Fatigue']

# 2. Ubah target 'Fatigue' menjadi angka (1 untuk 'Yes', 0 untuk 'No')
# Pastikan nilai di dalam .map() sesuai dengan nilai di kolom Anda!
try:
    y_encoded = y.map({'Yes': 1, 'No': 0})
except Exception as e:
    # Jika nilai Anda bukan 'Yes'/'No', ganti di sini. Misal: {'Positive':1, 'Negative':0}
    print(f"Error saat mapping target: {e}")
    print("Pastikan nilai di dalam y.map() sesuai dengan nilai unik di kolom 'Fatigue' Anda.")
    # Jika error, cek nilai unik dengan: print(y.unique())


# 3. Ubah semua fitur kategorikal menjadi angka dengan One-Hot Encoding
# pd.get_dummies() secara otomatis akan mengubah kolom teks dan membiarkan kolom angka
# drop_first=True untuk mencegah redundansi data yang bisa mengganggu beberapa model
X_encoded = pd.get_dummies(X, drop_first=True)

# Tampilkan 5 baris pertama dari data yang sudah siap
print("--- 5 Baris Pertama Fitur (X) Setelah Encoding ---")
print(X_encoded.head())

print("\n--- 5 Baris Pertama Target (y) Setelah Encoding ---")
print(y_encoded.head())
```

Memisahkan Data Training dan Testing
Kita akan membagi data menjadi dua: satu set untuk melatih model, dan satu set lagi untuk menguji performanya.

```python
from sklearn.model_selection import train_test_split

# Pisahkan data dengan rasio 80% untuk training dan 20% untuk testing
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

print(f"Ukuran data training (X_train): {X_train.shape}")
print(f"Ukuran data testing (X_test): {X_test.shape}")
```

Catatan: stratify=y_encoded adalah parameter penting yang memastikan proporsi target (jumlah 1 dan 0) di data training dan testing sama dengan proporsi di data asli, ini sangat baik untuk stabilitas model.

# Modelling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Buat objek model
# n_estimators adalah jumlah "pohon" dalam "hutan", random_state untuk hasil yang konsisten
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 2. Latih model dengan data training
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    roc_auc_score
)
import matplotlib.pyplot as plt

# 1. Lakukan prediksi pada data testing
y_pred = model.predict(X_test)

# --- Metrik yang Sudah Ada ---

# Akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi Model: {accuracy:.4f}")

# Presisi, Recall, F1-Score
print("\nLaporan Klasifikasi (Precision, Recall, F1-Score):")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues')
plt.show()


# --- Metrik Baru: AUC-ROC ---

# 2. Dapatkan probabilitas prediksi untuk kelas positif (kelas 1)
# Kurva ROC memerlukan probabilitas, bukan hanya hasil prediksi 0 atau 1
y_pred_proba = model.predict_proba(X_test)[:, 1]

# 3. Hitung AUC Score
auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"\nAUC Score: {auc_score:.4f}")

# 4. Hitung nilai untuk Kurva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# 5. Plot Kurva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='orange', label=f'ROC curve (area = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid()
plt.show()
```

# Model matematik


### **Dasar: Komponen Confusion Matrix**

Semua metrik ini berasal dari empat komponen dasar hasil prediksi yang bisa divisualisasikan dalam *Confusion Matrix*. Dalam konteks kasus Anda (target: `Fatigue`):

* **True Positive (TP):** Prediksi Benar Positif.
    * **Kasus:** Pasien sebenarnya `Fatigue`, dan model memprediksi `Fatigue`.
* **True Negative (TN):** Prediksi Benar Negatif.
    * **Kasus:** Pasien sebenarnya `Tidak Fatigue`, dan model memprediksi `Tidak Fatigue`.
* **False Positive (FP):** Prediksi Salah Positif (Error Tipe I).
    * **Kasus:** Pasien sebenarnya `Tidak Fatigue`, tetapi model memprediksi `Fatigue`. Ini adalah "alarm palsu".
* **False Negative (FN):** Prediksi Salah Negatif (Error Tipe II).
    * **Kasus:** Pasien sebenarnya `Fatigue`, tetapi model memprediksi `Tidak Fatigue`. Ini adalah "kasus yang terlewatkan".

Sekarang, mari kita lihat rumus dari setiap metrik.

---

### 1. Akurasi (Accuracy)

**Konsep:** Seberapa sering model membuat prediksi yang benar secara keseluruhan?

**Persamaan:**
$$\text{Akurasi} = \frac{\text{Jumlah Prediksi Benar}}{\text{Total Seluruh Prediksi}} = \frac{TP + TN}{TP + TN + FP + FN}$$

**Penjelasan:**
Rumus ini menjumlahkan semua prediksi yang benar (baik yang positif maupun negatif) dan membaginya dengan total seluruh data. Akurasi adalah metrik yang baik jika jumlah data untuk setiap kelas (misal, jumlah pasien `Fatigue` dan `Tidak Fatigue`) relatif seimbang.

---

### 2. Presisi (Precision)

**Konsep:** Dari semua prediksi `Fatigue` yang dibuat oleh model, berapa persen yang benar-benar `Fatigue`?

**Persamaan:**
$$\text{Presisi} = \frac{\text{Prediksi Benar Positif}}{\text{Total Prediksi Positif}} = \frac{TP}{TP + FP}$$

**Penjelasan:**
Presisi fokus pada kualitas prediksi positif. Metrik ini sangat penting ketika biaya dari *False Positive* (FP) tinggi. Contoh: Jika prediksi `Fatigue` akan memicu tindakan medis yang mahal, kita ingin prediksi tersebut sangat presisi untuk menghindari biaya pada pasien yang sehat.

---

### 3. Recall (Daya Ingat / Sensitivitas)

**Konsep:** Dari semua pasien yang sebenarnya `Fatigue`, berapa persen yang berhasil diidentifikasi oleh model?

**Persamaan:**
$$\text{Recall} = \frac{\text{Prediksi Benar Positif}}{\text{Total Aktual Positif}} = \frac{TP}{TP + FN}$$

**Penjelasan:**
Recall fokus pada kuantitas atau kelengkapan dalam menemukan semua kasus positif. Metrik ini sangat penting ketika biaya dari *False Negative* (FN) tinggi. Contoh: Dalam konteks medis, melewatkan pasien yang sebenarnya `Fatigue` (FN) bisa lebih berbahaya daripada memberikan alarm palsu (FP).

---

### 4. F1-Score

**Konsep:** Rata-rata harmonik (Harmonic Mean) dari Presisi dan Recall. Metrik ini mencari keseimbangan antara keduanya.

**Persamaan:**
$$\text{F1-Score} = 2 \times \frac{\text{Presisi} \times \text{Recall}}{\text{Presisi} + \text{Recall}}$$
Jika kita substitusi rumus Presisi dan Recall, persamaannya menjadi:
$$\text{F1-Score} = \frac{2 \times TP}{2 \times TP + FP + FN}$$

**Penjelasan:**
F1-Score sangat berguna ketika Anda membutuhkan keseimbangan antara Presisi dan Recall, terutama jika distribusi kelas tidak seimbang. Jika salah satu dari Presisi atau Recall sangat rendah, F1-Score juga akan rendah.

---

### 5. Kurva ROC dan AUC

Kurva ROC tidak memiliki satu "persamaan" tunggal, melainkan sebuah plot yang dibuat berdasarkan dua metrik turunan:

* **True Positive Rate (TPR):** Ini adalah nama lain dari **Recall**.
    $$\text{TPR (Recall)} = \frac{TP}{TP + FN}$$
* **False Positive Rate (FPR):** Dari semua yang sebenarnya `Tidak Fatigue`, berapa persen yang salah diprediksi sebagai `Fatigue`?
    $$\text{FPR} = \frac{\text{Salah Positif}}{\text{Total Aktual Negatif}} = \frac{FP}{FP + TN}$$

**Kurva ROC** adalah plot dari TPR (sumbu Y) melawan FPR (sumbu X) pada berbagai ambang batas klasifikasi.

**AUC (Area Under the Curve):**
AUC adalah luas area di bawah kurva ROC. Secara matematis, ini adalah nilai integral dari kurva tersebut.

**Persamaan Konseptual:**
$$\text{AUC} = \int_{0}^{1} \text{TPR}(\text{FPR}) \, d(\text{FPR})$$

**Penjelasan Sederhana:**
Anda tidak perlu menghitung integral ini secara manual karena library `scikit-learn` melakukannya untuk Anda. Makna dari **AUC** adalah probabilitas bahwa model Anda akan memberikan skor prediksi yang lebih tinggi untuk sampel positif acak (seorang pasien `Fatigue`) dibandingkan untuk sampel negatif acak (seorang pasien `Tidak Fatigue`). Nilai AUC 1.0 berarti pemisah yang sempurna, sementara 0.5 berarti setara dengan tebakan acak.
